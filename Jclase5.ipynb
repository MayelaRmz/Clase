```python
path = 'datasets/Canada.xlsx'
df = pd.read_excel(path, 
                   sheet_name = 'Canada by Citizenship', 
                   skiprows = range(20), 
                   skipfooter = 2, 
                   engine = 'openpyxl'
                  )
```

#### Make a comparitive study on top 5 countrys with maximum number of Immigrants during 1980 - 2013

```python
top = df.sort_values(by = 'Total', 
                      ascending = False, 
                      axis = 0)
top6 = top.head(6)
top6
```

```python
top6 = top6[years].transpose()
top6
```

```python
top6.plot(kind = 'line', 
          figsize = (15, 8)
         )
plt.title('A title', fontsize = 15)
plt.ylabel('# immingrants', fontsize = 15)
plt.xlabel('Years', fontsize = 15)
plt.show(block = True)
```

### Area plots (Staked line plot) 

```python
top6.plot(kind = 'area', 
          figsize = (15, 8), 
          stacked = False, # DO NOT aggragating the values
          alpha = 0.15     # intensity of the area in graph 
         )
plt.title('Title', fontsize = 18)
plt.show()
```
