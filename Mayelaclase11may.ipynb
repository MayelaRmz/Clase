```python
path = 'datasets/Cartwheeldata.csv'
df = pd.read_csv(path)
df.head()
```

```python
import seaborn as sns
```

```python
# Scatter plot 
sns.lmplot(x = 'Wingspan', 
          y = 'CWDistance',
          data = df,
          fit_reg = False)

plt.title('Scatterplot of CWDistance vs Winspan')
plt.show()
```

```python
sns.lmplot(x = 'Wingspan', 
          y = 'CWDistance',
          data = df,
          fit_reg = False,
          hue = 'Gender' ) ## <- Categorical data 

plt.title('Scatterplot of CWDistance vs Winspan')
plt.show()
```

```python
sns.swarmplot(x = 'Gender',
              y = 'CWDistance',
              data = df
             )
```

Box plot of gender = M fpr score

```python
sns.boxplot(data = df.loc[df['Gender']=='M',
                         ['Score']]
                         
           )
```

```python
fig = plt.figure()
ax1= fig.add_subplot(1,2,1) #1 row, 2 columns, position 1
sns.boxplot(data = df.loc[df['Gender']=='M',
                          ['Score']
                         ]
           )
```

```python
fig = plt.figure()
ax1= fig.add_subplot(1,2,1) #1 row, 2 columns, position 1
sns.boxplot(data = df.loc[df['Gender']=='M',
                          ['Score']
                         ]
           )
ax2= fig.add_subplot(1,3,3) #1 row, 2 columns, position 1
sns.boxplot(data = df.loc[df['Gender']=='F',
                          ['Score']
                         ]
           )
ax1.title.set_text('Scores of M')
ax2.title.set_text('Scores of F')
```
