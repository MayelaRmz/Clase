### Anthropometry Dataset

```python
# Libraries

import pandas as pd
import numpy as np
```

```python
# Reading the dataset

df = pd.read_csv('datasets/antropometria-dataset-2.csv')
df.head()
```

```python
# Creating a pivot table with the average weight of people by state

avgWeight = pd.pivot_table(df,
                           index = 'desc_ent',
                           values = 'peso',
                           aggfunc = 'mean'
                          )
avgWeight
```

```python
avgWeight = pd.pivot_table(df,
                           index = 'desc_ent',
                           values = 'peso',
                           aggfunc = 'median'
                          )
avgWeight
```

```python
# Creating a pivot table with the average weight of people by gender and state

avgWeight = pd.pivot_table(df,
                           index = 'desc_ent',
                           values = 'peso',
                           aggfunc = 'mean',
                           columns = 'sexo'
                          )
avgWeight
```

```python
avgWeight = pd.pivot_table(df,
                           index = 'desc_ent',
                           values = 'peso',
                           aggfunc = 'median',
                           columns = 'sexo'
                          )
avgWeight
```

```python
# Describing the pivot table

avgWeight.describe()
```

```python
# Describing the pivot table but grouping by state

avgWeight.groupby('desc_ent').describe()
```

##### Conclusions

Generally, the average weight of sex 2 is lower than sex 1. The lowest average weight for sex 1 is located in Chiapas and for sex 2 is located also in Chiapas. For the highest average weight for sex 1 is located in Baja California Sur and for sex 2 is located in Baja California Sur. Based on this dataset we can conclude that the lowest weights in average are in Chiapas and the highest weights are in Baja California Sur.

### Adult Dataset

```python
# Cleaned from the dataset the test_category_male and test_category_female rows
# Reading the dataset

df = pd.read_csv('datasets/adult.csv')
df.head()
```

```python
numMarital = pd.pivot_table(df,
                            index = ['education', 'marital-status'],
                            values = ['age', 'hours-per-week'],
                            aggfunc = 'mean',
                           )
numMarital
```

```python
numMarital = pd.pivot_table(df,
                            index = ['education', 'marital-status'],
                            values = ['age', 'hours-per-week'],
                            aggfunc = 'median',
                            columns = 'sex',
                           )
numMarital.head(1000000)
```

```python
# Creating the pivot table and generating a count and % column for marital status

numMarital = pd.pivot_table(df,
                            index = 'marital-status',
                            values = 'age',
                            aggfunc = 'count'
                           )
numMarital['count'] = numMarital['age']
del numMarital['age']
numMarital['%'] = numMarital['count']/sum(numMarital['count']) * 100
numMarital
```

```python
# Creating a new row 'Married' and summing the different types of marital status under married

numMarital.loc['Married'] = numMarital.iloc[1:4].sum()
numMarital
```

```python
# Dropping the previous marital status under married for clean-up

numMarital.drop(numMarital.index[1:4])
```
