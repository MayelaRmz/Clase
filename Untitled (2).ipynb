```python
path = 'datasets/Cartwheeldata.csv'
df = pd.read_csv(path)
df.head()
```

```python
import seaborn as sns
```

```python
# Scatter plot 
sns.lmplot(x = 'Wingspan', 
          y = 'CWDistance',
          data = df,
          fit_reg = False)

plt.title('Scatterplot of CWDistance vs Winspan')
plt.show()
```

```python
sns.lmplot(x = 'Wingspan', 
          y = 'CWDistance',
          data = df,
          fit_reg = False,
          hue = 'Gender' ) ## <- Categorical data 

plt.title('Scatterplot of CWDistance vs Winspan')
plt.show()
```

```python
sns.swarmplot(x = 'Gender',
              y = 'CWDistance',
              data = df
             )
```
